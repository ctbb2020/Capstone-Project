```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

<b>Part 1: Scrape the wiki page and create the clean-up dataframe

```python
#Scrape data from wiki site
List_url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = requests.get(List_url).text
soup = BeautifulSoup(source, 'xml')
table=soup.find('table')
```

```python
#Create a dataframe which consists of 3 columns
column_names = ['PostalCode','Borough','Neighborhood']
df = pd.DataFrame(columns = column_names)
```

```python
for tr_cell in table.find_all('tr'):
    row_data=[]
    for td_cell in tr_cell.find_all('td'):
        row_data.append(td_cell.text.strip())
    if len(row_data)==3:
        df.loc[len(df)] = row_data
df.head()
```

```python
#Data clean-up
df1 = df[df.Borough != 'Not assigned']
df2 = df1.groupby(['PostalCode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)
df2['Neighborhood'] = np.where(df2['Neighborhood'] == 'Not assigned',df2['Borough'], df2['Neighborhood'])
df2
```

```python
df2.shape
```

<b>Part 2: Add the latitudes and longitudes to the above datafrome table 

```python
geo = pd.read_csv('https://cocl.us/Geospatial_data')
geo.head()
```

```python
geo.rename(columns={'Postal Code':'PostalCode'},inplace=True)
df3 = pd.merge(df2,geo,on='PostalCode')
df3
```

<b> Part 3: Explore and cluster the neighborhoods in Toronto

```python
#Get an updated dataframe which consists with only Toronto in 'Borough' column
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

```python
#Use folium to create a Toronto map using the above new dataframe
import folium
toronto_map = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighborhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='red',
    fill=True,
    fill_color='#31cc77',
    fill_opacity=0.7,
    parse_html=False).add_to(toronto_map)
toronto_map
```

```python
#Cluster the neighborhoods using K-means
from sklearn.cluster import KMeans

k=5
toronto_clustering = df4.drop(['PostalCode','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df4
```

```python
#Create an updated map with the above dataframe
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# Choose colors scheme for clusters
import matplotlib.cm as cm
import matplotlib.colors as colors

x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# Add markers to map
markers_colors = []
for lat, lon, neighborhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighborhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
